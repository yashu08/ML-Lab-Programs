```python
import matplotlib.pyplot as plt

days = ["mon", "tue", "wed", "thu", "fri"]
temp= [10, 15, 20, 25, 30]

plt.scatter(days, temp, c="blue")
plt.title('Temperature Variation', fontsize=10)
plt.ylabel('Temperature (°C)')
plt.xlabel('Months')
plt.grid(False) 
plt.show()
```

```python
#bar graph
plt.bar(days, temp, color ='pink', width = 0.2)
plt.xlabel("days")
plt.ylabel("Temp (°C)")
plt.title("Temperature Variation")
plt.show()
```

```python
#line chart
plt.plot(days, temp) 
plt.title("Temperature Variation")
plt.xlabel("Temperature (°C)")
plt.ylabel("Days")
plt.show()
```

```python
#pie chart
import pandas as pd
data = {'categories' :['Chocolate', 'Chips', 'Cake', 'Sweets'],
'sales_amount': [4000, 9500, 3500, 8000]}
df = pd.DataFrame(data)
plt.pie(df['sales_amount'], labels=df['categories'])
plt.title('Sales Distribution by Category')
plt.show()
```
