```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X = data[['TV', 'Radio', 'Newspaper']]
y = data['Sales']
```

```python
x_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
multiple_regressor = LinearRegression()
multiple_regressor.fit(x_train, y_train)
```

```python
print('Intercept: ', multiple_regressor.intercept_)
print('Coefficients:', multiple_regressor.coef_)
```

```python
print('Regression Equation: Sales = 6.976 + 0.6759 *TV')
```

```python
y_pred_test = multiple_regressor.predict(X_test)
y_pred_train = multiple_regressor.predict(X_train)
```
